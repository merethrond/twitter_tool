### See if the followers info can be retrieved via paging( Result - not feasible time wise)

```python
count = 0
id_list = []
followers_count_list = []
friends_count_list = []
statuses_count_list = []
created_at_list = []
location_list = []


for page in tweepy.Cursor(api_dict['youthfoundationss@gmail.com'].followers, 'carlahome80', cursor = -1).pages(2):
    for follower in page:
        id_list.append(follower.id)
        followers_count_list.append(follower.followers_count)
        friends_count_list.append(follower.friends_count)
        statuses_count_list.append(follower.statuses_count)
        created_at_list.append(follower.created_at)
        location_list.append(follower.location)
        count += 1
    print(count)
    time.sleep(60)
```

```python
# Stacking the lists column wise, so that they can fit in shape in the dataframe.
df_details = pd.DataFrame(np.column_stack([id_list, followers_count_list, friends_count_list, statuses_count_list, created_at_list, location_list]),\
                 columns = ['id', 'followers_count', 'friends_count', 'statuses_count', 'created_at', 'location'])
```
